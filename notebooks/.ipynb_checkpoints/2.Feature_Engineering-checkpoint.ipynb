```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis
import os
```

### 1. SP500 Tickers

```python
folder_path = "../data/processed/sp500/"
combined_df = None

for filename in os.listdir(folder_path):
    if filename.endswith(".csv"):
        ticker = filename.replace(".csv", "") 
        filepath = os.path.join(folder_path, filename)

        df = pd.read_csv(filepath, usecols=["Price", "Close"])
        df.rename(columns={"Close": ticker}, inplace=True)
        df.set_index("Price", inplace=True)

        if combined_df is None:
            combined_df = df
        else:
            combined_df = combined_df.join(df, how='inner')

combined_df.index.name = "Date"
combined_df.columns.name = None
combined_df = combined_df.drop(combined_df.index[0])
```

```python
log_returns = np.log(combined_df.astype(float) / combined_df.astype(float).shift(1)).dropna()
```

```python
log_returns.tail()
```

```python
rolling_volatility = log_returns.rolling(window=30).std().dropna()
```

```python
rolling_volatility.head()
```

```python
combined_df = combined_df.astype(float)
peak = combined_df.cummax()
drawdowns = (combined_df - peak) / peak 
```

```python
drawdowns
```

```python
cumulative = (log_returns + 1).cumprod()

running_max = cumulative.cummax()
drawdown = (cumulative - running_max) / running_max
max_drawdown = drawdown.min()
```

```python
avg_corr = log_returns.corr().mean()
```

```python
avg_corr
```

```python
sharpe_ratio = log_returns.mean()/(log_returns.std()) * np.sqrt(252)
```

```python
max_drawdown
```

```python
sharpe_ratio
```

```python
vol = log_returns.std().dropna()
```

```python
vol
```

```python
skewness = log_returns.apply(skew, axis=0)
kurt = log_returns.apply(kurtosis, axis=0)
```

```python
kurt
```

```python
summary_df = pd.DataFrame({
    "Avg Correlation": avg_corr,
    "Volatility": vol,
    "Max Drawdown": max_drawdown,
    "Skewness": skewness,
    "Kurtosis": kurt,
    "Sharpe Ratio": sharpe_ratio
    
})

summary_df = summary_df.round(4)
summary_df = summary_df.sort_index()
```

```python
summary_df
```

```python
summary_df.columns = summary_df.columns.astype(str).str.strip()
features = summary_df.columns.tolist()
```

```python
for col in features:
    plt.figure(figsize=(6, 4))
    sns.histplot(summary_df[col], kde=True, bins=10)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
for col in features:
    plt.figure(figsize=(5, 1.5))
    sns.boxplot(x=summary_df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(summary_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
for col in features:
    plt.figure(figsize=(6, 4))
    sns.kdeplot(summary_df[col], fill=True)
    plt.title(f"KDE Plot of {col}")
    plt.show()
```
